# 1. RADseq

Секвенирование ДНК, ассоциированное с сайтом рестрикции, стало мощным инструментом для генетического картирования и анализа локусов количественных признаков, адаптации и филогеографии. В основу метода положено разрезание ДНК в определенных участках - сайтах рестрикции посредством одного или нескольких ферментов - рестриктаз. Результаты рестрикции могут быть отфильтрованы по длине участка, или не отфильтрованы. 

<img src="RAD.jpg" width ="700" height="700" style="float:left"/>

Основным преимуществом RAD-seq является возможность одновременного анализа десятков тысяч генетических локусов со значительно меньшими затратами по сравнению с подходами полногеномного секвенирования. RAD-seq может использоваться при скрещивании любого дизайна, а также диких популяций и организмов с референсным геномом или без него. Последующий анализ включает обнаружение SNP, картирование сцепления и картирование QTL (Quantitative Trait Loci), а также более сложные анализы, такие как популяционные геномные и филогеографические исследования.

Существует несколько методик RAD-seq, которые различаются числом рестриктаз, методом присоединения адаптеров и наличием/отсутствием этапа отбора длины ДНК фрагментов (рисунок ниже). Все методы начинаются с относительно высокомолекулярной геномной ДНК и начинаются с ее обработки одним или несколькими рестрикционными ферментами. Все подходы используют специфические адаптеры секвенирования или двухцепочечные олигонуклеотиды, которые требуются для всех платформ NGS секвенирования. Адаптеры, добавленные в протоколах RADseq, могут содержать баркод-адаптеры, которые используются для идентификации отдельных образцов, которые затем будут разделены методом демультиплексирования. В зависимости от используемого фермента(-ов) протоколы RADseq также могут регулировать размеры фрагментов ДНК, которые являются оптимальными для секвенирования следующего поколения. 

<img src="andrews2016-4.1.jpg" width ="700" height="600">

- Классический метод **RAD-seq** отличается использованием одной рестриктазы, после действия которой лигируются адаптеры для секвенирования с обоих концов ДНК. Затем, цепи ДНК повторно разрезаются для достижения оптимальной длины ДНК для NGS секвенирования. Затем, прикрепляются Y адаптеры, используемые для платформы Illumina, после чего следует ПЦР.
- **2bRAD** отличается от остальных методов тем, что рестриктаза разрезает ДНК с обоих сторон от сайта узнавания так, что длина ДНК вставки составляет всего 33–36 п.н. Оба метода объединяются между собой наличием в реакции всего одной рестриктазы.

<img src="andrews2016-4.2.jpg" width ="1000" height="900">

Следующие три метода используют несколько рестркционных ферментов. 
- Для генотипирования путем секвенирования (Genotyping by sequencing, **GBS**) используется фермент для разрезания ДНК на большие фрагменты, после чего следует фермент, разрезающий ДНК на более мелкие фрагменты. ПЦР также отбирает только короткие фрагменты, используемые для секвенирования. 
- **ezRAD** использует один или несколько обычных режущих ферментов, после чего используется запатентованный набор для подготовки библиотеки Illumina. После лигирования адаптеров обязательно следует этап отбора длины фрагментов методами магнитных частиц или фильтрование в агарозном геле. 
- Наконец, метод RAD с двойным разрезанием и отбором длины (double digest RAD **ddRAD**) использует два рестрикционных фермента, специфичных для различных сайтов узнавания. После этого следует лигирование адаптеров для любой платформы секвенирования, после чего происходит отбор участков ДНК определенной длины (size-select) с помощью агарозного геля или магнитных частиц.

# 2. Stacks

<img src="stacks_logo.png" width ="200" height="200" align="left"/>

программа для построения локусов из коротких ридов. Stacks была разработана для работы с данными на основе  рестрикционных ферментов, такими как RAD-seq, с целью построения генетических карт и проведения популяционной  геномики и филогенетики. Написан на C++.

<a href="https://catchenlab.life.illinois.edu/stacks/">https://catchenlab.life.illinois.edu/stacks/</a>

**ustacks** (unique stacks) - принимает в качестве входных данных риды и выравнивает их в точно совпадающие стеки (*от англ. stack-стопка*) или предполагаемые аллели. Сравнивая стеки, ustacks формирует набор предполагаемых локусов и обнаруживает SNP в каждом локусе, используя структуру максимального правдоподобия.<br>
**cstacks** (catalog of stacks) - создает каталог локусов <br>
**sstacks** (sets of stacks) - сопоставляет выборки с каталогом<br>
**tsv2bam** - транспонирует данные так, чтобы они были ориентированы по локусу, а не по выборке, а также добавляет обратные риды при их наличии<br>
**gstacks** - идентифицирует SNP для каждого локуса, а затем генотипирует каждую особь по каждому идентифицированному SNP. После того, как SNP были идентифицированы и генотипированы, gstacks будет фазировать SNP в каждом локусе, в каждом индивидууме, в набор гаплотипов. Программа gstacks способна удалять дубликаты PCR (пары чтений с одинаковой длиной вставки), если требуется.<br>
**populations** - анализирует популяцию отдельных образцов, вычисляя ряд статистических данных популяционной генетики, а также экспортируя множество стандартных выходных форматов. Карта популяции с указанием того, какие люди к какой популяции принадлежат, передается в программу, и затем вычисляется статистика популяционной генетики - ожидаемая / наблюдаемая гетерозиготность, $π$ и $F_{is}$ в каждой позиции нуклеотида. Эта программа популяций сравнивает все популяции попарно для вычисления $F_{st}$. Также может вычисляет ряд статистических данных популяционной генетики на основе гаплотипов, включая разнообразие гаплотипов, $Φ_{st}$ и $F_{st}$ ’.<br>
Программа populations предоставляет возможности для сильной фильтрации, чтобы включать только локусы или варианты сайтов, которые встречаются с определенной частотой в каждой популяции или в метапопуляции.
Выдает philyp файл.

- Начальный этап алгоритма сборки **ustacks** *de novo* формирует совпадающие стеки из необработанных чтений.


Параметр минимальной глубины стека контролирует количество необработанных чтений, необходимых для формирования начального стека. Если глубина покрытия для определенного стека ниже заданного значения, то аллель не будет сформирован, и эти чтения временно откладываются алгоритмом и не помещаются в стек. Необработанные чтения, помещенные в стек, называются первичными, а отложенные - вторичными.

<img src="min_depth.png" width ="600" height="600" style="float:center"/>

___

- Консенсусная последовательность из каждого стека разбивается на kmers на втором этапе алгоритма **ustacks** и сохраняется в словаре. Два стека, которые имеют определенное количество общих k-мер, рассматриваются как потенциально совпадающие и выровниваются вместе. Если количество нуклеотидных несовпадений меньше, чем расстояние между стеками, стеки объединяются в локус.

<img src="stack_dist.png" width ="500" height="600" style="float:center"/>

___

- Как только локусы сформированы, вторичные чтения возвращаются в анализ и выравниваются по собранным локусам с использованием более допустимого значения нуклеотидного несоответствия (вы можете контролировать это значение). Этот процесс обеспечивает большую глубину, которая помогает модели вызова SNP в обнаружении полиморфизмов. Локус с одним полиморфизмом показан на рисунке.

# 3. Использование Stacks

Для дальнейшей работы воспользуемся программой Stacks, предназначенной для обработки RAD-seq данных.
Для установки программы используйте следующие команды:
<br>**wget https://catchenlab.life.illinois.edu/stacks/source/stacks-2.60.tar.gz**
<br>**tar xfvz stacks-2.60.tar.gz**
<br>**cd stacks-2.60**
<br>**./configure**
<br>**make**
<br>**sudo make install**

Для скачивания SRA (Sequence Read Archive) необходимо также установить SRA-Toolkit, выполнив ряд команд:
<br>**wget https://ftp-trace.ncbi.nlm.nih.gov/sra/sdk/2.11.3/sratoolkit.2.11.3-ubuntu64.tar.gz**
<br>**tar xzvf sratoolkit.2.11.3-ubuntu64.tar.gz**
<br>**export PATH=$PATH:/директория/sratoolkit.2.11.3-ubuntu64/bin**

Используйте команду prefetch для скачивания пяти файлов:
**<br>SRR11790139 
<br>SRR11790141
<br>SRR11790147
<br>SRR11790148
<br>SRR11790152**
<br>Данные последовательности принадлежат ручейникам - амфибиотическим насекомым, личиночная стадия которых проходит в водной среде, тогда как взрослые особи (имаго) вылетают после стадии куколки. Копуляция имагинальных пар происходит в водную среду. Личинки и имаго ручейников являются кормовой базой для лососевых рыб, включая хариусов :)

### Задание № 3.4
  - Сколько локусов содержал совокупный для всей выборки каталог?

### Ответ:

***

Выполните сравнение каталогов локусов каждой особи со всей выборкой посредством команды **sstacks**: 
<br>**sstacks -P "директория с файлами после cstacks" -M "Путь до файла popmap" -p 8**

Выполните транспонирование данных по локусам, а не выборке посредством **tsv2bam**. Свяжите парные риды для каждой особи.
<br>**tsv2bam -P "директория с файлами после sstacks" -M "Путь до файла popmap" --pe-reads-dir "путь до обратных ридов" -t 8**

Выполните анализ каждой особи внутри ридов с помощью gstacks:
<br>**gstacks -P "директория с файлами после tsv2bam" -M "Путь до файла popmap" -t 8**

Наконец, используйте команду **populations** для вывода результатов анализа. В данном случае, нам необходимо получить **Phylip** файл для построения филогенетического дерева
<br>**populations -P "директория с файлами после gstacks" -M "Путь до файла popmap" --phylip-var-all -t 8**

Откройте полученный файл и удалите лишние строки в конце.

# 4. Установка и запуск программы для построения дерева методом максимального правдоподобия

1. Установите на компьютер программу RAxML:<br>
<strong>sudo apt-get update -y</strong><br>
<strong>sudo apt-get install -y raxml</strong>

2. Зайдите в терминал. Находясь в директории с файлом For_RAxML.phylip, запустите программу RAxML:<br>
<strong>raxmlHPC -T 14 -s For_RAxML.phylip -m GTRGAMMA -p 1234 -n trees -# 500</strong>

3. После построения 500 деревьев напишите команду для расчета bootstrap поддержек:<br>
<strong>raxmlHPC -T 14 -s For_RAxML.phylip -m GTRGAMMA -p 1234 -b 1234 -n bootstrap -# 500</strong>

4. Для объединения деревьев и bootstrap поддержек и получения финального дерева напишите:<br>
<strong>raxmlHPC -T 14 -s For_RAxML.phylip -m GTRGAMMA -f b -t RAxML_bestTree.trees -z RAxML_bootstrap.bootstrap -n RAxML_tree</strong>

# 5. Отрисовка дерева в программе Figtree

1. Установите программу Figtree:<br>
<strong>sudo apt-get update -y</strong><br>
<strong>sudo apt-get install -y figtree</strong>

2. Запустите программу командой <strong>figtree</strong>

3. Откройте файл с названием:<br>
<strong>RAxML_bipartitions.RAxML_tree</strong>

4. Приведите дерево в более читаемый вид путем внесения изменений по следующим параметрам:<br>
<strong>Trees - Root tree - Midpoint</strong><br>
<strong>Order nodes - increasing</strong><br>
<strong>Tip Labels - FontSize</strong> (измените шрифт по желанию)<br>
<strong>Node Labels - Display - labels</strong> (измените поддержки узлов по желанию)<br>

   Для вывода внешней группы в качестве корневой необходимо найти ее на дереве и нажать <strong>Reroot</strong>
